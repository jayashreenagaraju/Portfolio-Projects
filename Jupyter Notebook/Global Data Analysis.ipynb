```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# load the dataset from a csv file
df = pd.read_csv(r"C:\Users\Lenovo\Desktop\Kaggle Data\world-data-2023.csv")
```

```python
df = df[['Country','Birth Rate', 'Fertility Rate','Infant mortality','Life expectancy','Physicians per thousand', 'Population']]
df
```

```python
df.info()
```

```python
# Converting 'Population' to strings before replacing commas
df['Population'] = df['Population'].astype(str).str.replace(',','')
df['Population']
```

```python
# Convert 'Population' to numeric with 'coerce' option
# 'Coerce' means - It convert all values in that column to numeric types. If any value cannot be converted (for instance, if there are non-numeric characters or strings that can't be interpreted as numbers), 
# then those specific values will be replaced with NaN(Not a number). 
df['Population'] = pd.to_numeric(df['Population'],errors = 'coerce')
```

```python
# Columns to impute
columns_to_impute = ['Birth Rate', 'Fertility Rate','Infant mortality','Life expectancy','Physicians per thousand', 'Population']
```

```python
# Impute missing values with column meadians
median_values = df[columns_to_impute].median(numeric_only=True)
df[columns_to_impute] = df[columns_to_impute].fillna(median_values)
```

```python
# Re-name columns for consistency
df.rename(columns = {'Birth Rate':'Birth_Rate'},inplace=True)
df.rename(columns = {'Fertility Rate':'Fertility_Rate'},inplace=True)
df.rename(columns = {'Infant mortality':'Infant_mortality'},inplace=True)
df.rename(columns = {'Life expectancy':'Life_expectancy'},inplace=True)
df.rename(columns = {'Physicians per thousand':'Physicians_per_thousand'},inplace=True)
df.columns
```

Visualisations

```python
# Display pair plots to explore correlations across the df
sns.pairplot(df)
```

```python
# Calculate and display correlation matrix heatmap

correlation_matrix =df.corr(numeric_only=True)
plt.figure(figsize=(10,8))
sns.heatmap(correlation_matrix,annot = True, cmap = 'coolwarm')
plt.title('Correlation Matrix')
```

```python
# Display distribution histograms for column

plt.figure(figsize=(12,8))

for column in df.columns[1:]:
    plt.subplot(2,3,df.columns.get_loc(column))
    sns.histplot(df[column],kde = True)
    plt.title(f'Distribution of {column}')
plt.tight_layout()
```

```python
#Display Scatter plot of Birth_Rate  vs Life_Expectancy
plt.figure(figsize=(12,8))

sns.regplot(data=df,x ='Birth_Rate', y ='Life_expectancy')
plt.title('Birth_Rate  vs Life_Expectancy')
```

```python
# Display top 10 countries with highest population
top_population_countries = df.nlargest(10,'Population')
```

```python
# Create a bar plot using seaborn
sns.set(style='whitegrid')
plt.figure(figsize=(12,6))
sns.barplot(x='Country', y = 'Population',data = top_population_countries,palette = 'bright')
plt.xticks(rotation=45,ha='right')
plt.title('Top 10 Countries with Highest Population')
```

```python
# Calculate and display average fertility rates by country
Avg_fertility_by_country=df.groupby('Country')['Fertility_Rate'].mean()
print("Average Fertility Rate by Country")
Avg_fertility_by_country
```

```python
# Calculate and display countries with the highest life expectancy
Highest = df.groupby('Country')['Life_expectancy'].max().nlargest(10)
print(Highest)
```

```python
# Calculate and display countries with the lowest life expectancy
Lowest = df.groupby('Country')['Life_expectancy'].min().nsmallest(10)
print(Lowest)
```
